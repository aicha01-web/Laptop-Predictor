**1-Importer les données depuis kaggle et visualiser**

```python
import numpy as np
import pandas as pd
```

```python
#lecture du ficier csv contenant les données
data = pd.read_csv("laptop_data.csv")
#affichage des premières lignes de données
data.head()
```

```python
data.info()
```

**Nettoyage des données**

```python
#Suppression de la colonne unnamed
data.drop(columns=['Unnamed: 0'],inplace=True)
data.info()
```

```python
#Vérification des donées dupliquées
data.duplicated().sum()
```

```python
#Suppression des valeurs dupliquées
# d2 = data.duplicated().drop_duplicates(keep="first",inplace=False)
# d2.head()
# df = data.loc[:,~data.columns.duplicated()].copy()
# data.duplicated().sum()
data = data.drop_duplicates(keep='first',inplace=False)
data.duplicated().sum()
# data.head()
# data.duplicated().sum()
```

```python
#Taille du dataset après nettoyage
data.__len__()
```

```python
#on cherche s'il y'a des valeurs nulls
data.isnull().sum()
```

```python
data['Ram'] = data['Ram'].str.replace('GB','')
data['Weight'] = data['Weight'].str.replace('kg','')
#conversion du prix en CFA
data['Price'] = data['Price'] * 7.34

data.head()
```

```python
#changement des types de variables pour le poids et la ram
data["Ram"]=data["Ram"].astype("int")
data["Weight"]=data["Weight"].astype("float")

data.info()
```

```python
data['Gpu'] = data['Gpu'].apply(lambda x : x.split()[0])
```

```python
data['CpuName'] = data['Cpu'].apply(lambda x :" " .join(x.split()[:3]))
def fetch_processor(text):
    if text == 'Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD Processor'
data['Cpu brand'] = data['CpuName'].apply(fetch_processor)
```

```python
data['Memory'] = data['Memory'].astype(str).replace('\.0', '', regex=True)
data["Memory"] = data["Memory"].str.replace('GB', '')
data["Memory"] = data["Memory"].str.replace('TB', '000')
```

```python
data.head()
```

```python

data.drop(columns = ['Cpu', 'CpuName'], inplace = True)
```

```python

data.head()
```

```python
#diviser les chaines avant et apres le plus
new2 = data["Memory"].str.split("+", n = 1, expand = True)
data["first"]= new2[0]
#enleve les espaces blancs
data["first"]=data["first"].str.strip()
data["second"]= new2[1]
```

```python
data["Layer1HDD"] = data["first"].apply(lambda x: 1 if "HDD" in x else 0)
data["Layer1SSD"] = data["first"].apply(lambda x: 1 if "SSD" in x else 0)
data["Layer1Hybrid"] = data["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
data["Layer1Flash_Storage"] = data["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)
```

```python
data['first'].values
```

```python
#Pour remplacer les valeurs non numériques par une chaine vide
data['first'] = data['first'].str.replace(r'\D', '')
#Remplace les NaN par 0
data["second"].fillna("0", inplace = True)
data["Layer2HDD"] = data["second"].apply(lambda x: 1 if "HDD" in x else 0)
data["Layer2SSD"] = data["second"].apply(lambda x: 1 if "SSD" in x else 0)
data["Layer2Hybrid"] = data["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
data["Layer2Flash_Storage"] = data["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)
```

```python
data['second'] = data['second'].str.replace(r'\D', '')
data["first"] = data["first"].astype(int)
data["second"] = data["second"].astype(int)
```

```python
data["Total_Memory"]=(data["first"]*(data["Layer1HDD"]+data["Layer1SSD"]+data["Layer1Hybrid"]+data["Layer1Flash_Storage"])+
                    data["second"]*(data["Layer2HDD"]+data["Layer2SSD"]+data["Layer2Hybrid"]+data["Layer2Flash_Storage"]))
data["Memory"]=data["Total_Memory"]
data["HDD"]=(data["first"]*data["Layer1HDD"]+data["second"]*data["Layer2HDD"])
data["SSD"]=(data["first"]*data["Layer1SSD"]+data["second"]*data["Layer2SSD"])
data["Hybrid"]=(data["first"]*data["Layer1Hybrid"]+data["second"]*data["Layer2Hybrid"])
data["Flash_Storage"]=(data["first"]*data["Layer1Flash_Storage"]+data["second"]*data["Layer2Flash_Storage"])
```

```python
data.head()
```

```python
new = data["ScreenResolution"].str.split("x", n = 1, expand = True) 
data["X_res"]= new[0]
data["Y_res"]= new[1]
data["Y_res"]= pd.to_numeric(data["Y_res"])
data["Y_res"]= data["Y_res"].astype(float)
#remplace les virgules par une chaine vide dans les X_res ensuite cherche les valeurs numeriques décimaux et enfin remplace par la moyenne
data["X_res"]=(data['X_res'].str.replace(',','').str.findall(r'(\d+\.?\d+)').apply(lambda x: pd.Series(x).astype(int)).mean(1))
data["X_res"]=pd.to_numeric(data["X_res"])
#pixel per inch
data["PPI"]=(((data["X_res"]**2+data["Y_res"]**2)**(1/2))/data["Inches"]).astype(float)
data["ScreenResolution"]=(data["X_res"]*data["Y_res"]).astype(float)
```

```python
data.head()
```

```python
data=data.drop(['first','second','Layer1HDD','Layer1SSD','Layer1Hybrid','Layer1Flash_Storage','Layer2HDD','Layer2SSD','Layer2Hybrid',
                'Layer2Flash_Storage','Total_Memory'],axis=1)
```

```python
data.head()
```

**Diviser en données d'entrainements et tests**

```python
from sklearn.model_selection import train_test_split

X = data.drop(columns=['Price'])
y = data['Price']

X
```

```python
y
```

```python
#Division en données d'entrainement et de test

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
X_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score,recall_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first',handle_unknown='ignore'),[0,1,6,7,9])
],remainder='passthrough')

step2 = LinearRegression()

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

real_error = [true - pred for true, pred in zip(y_test, y_pred)]
print('Real error',real_error)
print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred) )
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error,r2_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first',handle_unknown='ignore'),[0,1,6,7,9])
],remainder='passthrough')

step2 = DecisionTreeRegressor()

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

real_error = [true - pred for true, pred in zip(y_test, y_pred)]
print('Real error',real_error)
print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred) )
```

```python
from sklearn.neighbors import KNeighborsRegressor

step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first',handle_unknown='ignore'),[0,1,6,7,9])
],remainder='passthrough')

step2 = KNeighborsRegressor()

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

real_error = [true - pred for true, pred in zip(y_test, y_pred)]
print('Real error',real_error)
print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred) )
```

```python
from xgboost import XGBRegressor

step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first',handle_unknown='ignore'),[0,1,6,7,9])
],remainder='passthrough')

step2 = XGBRegressor(n_estimators=45,max_depth=5,learning_rate=0.5)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

real_error = [true - pred for true, pred in zip(y_test, y_pred)]
print('Real error',real_error)
print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor


step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse=False,drop='first',handle_unknown='ignore'),[0,1,6,7,9])
],remainder='passthrough')

step2 = RandomForestRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15)

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

real_error = [true - pred for true, pred in zip(y_test, y_pred)]
print('Real error',real_error)
print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_absolute_error(y_test,y_pred))
```
